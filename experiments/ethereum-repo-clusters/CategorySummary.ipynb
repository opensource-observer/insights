```python
import json
import pandas as pd
```

```python
df = pd.read_parquet('output/ethereum_repos_unified.parquet')
df["categorizations_list"] = df["categorizations_json"].apply(json.loads)

def persona_to_category_map(cats_list):
    return { d["persona_name"]: d["category"] for d in cats_list }
df_persona_map = df["categorizations_list"].apply(persona_to_category_map)
df_persona_cols = pd.json_normalize(df_persona_map)

df = df.join(df_persona_cols)
df = df.drop(columns=["categorizations_list", "categorizations_json", "readme_md"])
#df.to_csv('categorizations.csv')
df
```
