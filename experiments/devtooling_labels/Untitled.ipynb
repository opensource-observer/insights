```python
df1 = (
    pd.read_parquet('output/devtooling_summarized.parquet')
    .drop(columns=['readme_md'])
    .set_index('repo_artifact_id')
)
df2 = (
    pd.read_csv('output/devtooling_w_categories.csv')
    .set_index('repo_artifact_id')
    [['security_veteran_tag', 'quick_shipper_tag', 'superchain_builder_tag']]
)
df = df1.join(df2).drop_duplicates()
```

```python
df[df['display_name'] == 'Remix Project']
```

```python
tag_cols = ['security_veteran_tag','quick_shipper_tag','superchain_builder_tag']
project_summaries = (
    df
    .groupby('project_id')[tag_cols]
    .apply(lambda g: dict(Counter(g.values.ravel())))
)
df['project_tag_summary'] = df['project_id'].map(project_summaries)
```

```python
agg = (
    df
    .groupby(['atlas_id','display_name'])
    .agg(
        repo_artifact_namespace = ('repo_artifact_namespace', lambda x: list(x.unique())),
        star_count               = ('star_count', 'sum'),
        fork_count               = ('fork_count', 'sum'),
        num_packages_in_deps_dev = ('num_packages_in_deps_dev', 'sum'),
        created_at               = ('created_at', 'min'),
        updated_at               = ('updated_at', 'max'),
        project_tag_summary      = ('project_tag_summary', 'first'),
    )
    .reset_index()
)
tag_counts = pd.json_normalize(agg['project_tag_summary']).fillna(0)
tag_props = tag_counts.div(tag_counts.sum(axis=1), axis=0)
final = pd.concat([
    agg.drop(columns=['project_tag_summary']),
    tag_props
], axis=1)

tag_cols = tag_props.columns.tolist()  # or explicitly list your tag columns

# 1. compute the max value and its corresponding tag for each row
max_vals = final[tag_cols].max(axis=1)
max_tags = final[tag_cols].idxmax(axis=1)

# 2. apply the 0.67 threshold
final['recommendation'] = np.where(
    max_vals > 0.66,
    max_tags,
    'Other'
)

final.to_csv('output/devtooling_consolidated.csv')
```
