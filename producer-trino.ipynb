```python
import pandas as pd
from sqlalchemy import create_engine
from trino.auth import BasicAuthentication

class Client:
    def __init__(
        self,
        user: str,
        host: str,
        port: int = 8080,
        catalog: str = "iceberg",
        schema: str = "oso",
        password: str = "",
        protocol: str = "http"
    ):
        self.engine = create_engine(
            f"trino://{user}@{host}:{port}/{catalog}/{schema}?protocol={protocol}",
            connect_args={"auth": BasicAuthentication(user, password)}
        )

    def to_pandas(self, query: str) -> pd.DataFrame:
        with self.engine.connect() as con:
            return pd.read_sql(query, con)

client = Client(
    user="producer-user",
    host="production-trino-production-trino-trino.tail7ec788.ts.net",
    catalog="iceberg",
    schema="oso"
)            
```
