```python
import pandas as pd
```

```python
METRICS_PATH = 'https://raw.githubusercontent.com/ethereum-optimism/Retro-Funding/refs/heads/main/eval-algos/S7/data/M1/onchain/onchain_metrics_by_project.csv'
RESULTS_PATH = 'https://raw.githubusercontent.com/ethereum-optimism/Retro-Funding/refs/heads/main/results/S7/m1_onchain_builders_results.json'
```

```python
df_metrics = pd.read_csv(METRICS_PATH)
df_results = pd.read_json(RESULTS_PATH)
```

```python
metric = 'contract_invocations_monthly'
period = 'Feb 2025'
```

```python
df = (
    df_results
    .set_index(['project_name', 'display_name'])[['op_reward']]
    .join(
        df_metrics.query(
            "metric_name == @metric and measurement_period == @period"
        ).pivot_table(
            index=['project_name', 'display_name'],
            columns='chain',
            values='amount',
            aggfunc='count'
        )
    )
    .fillna(0)
    .sort_values(by='op_reward', ascending=False)
)
df
```

```python
df.to_csv("data/M1_onchain_builders_by_chain.csv")
```
