```python
import pandas as pd
```

```python
df = pd.read_parquet('data/enriched_tvl_events.parquet')
df['amount'] = df['amount'].apply(int)
df.tail(1)
```

```python
df = df[df['event_type'] == 'TOKENS_IN_USD']
df
```

```python
df[df['to_artifact_name'] == 'aave-v3'].sort_values(by='time').tail()
```

```python
df_tvl = df.groupby(['to_artifact_name', 'to_artifact_namespace', 'time'], as_index=False)['amount'].sum()
```

```python
df_tvl[df_tvl['to_artifact_name'] == 'aave-v3'].sort_values(by='time')
```

```python
op = (
    df[(df['to_artifact_name'] == 'aave-v3')
       & (df['to_artifact_namespace'] == 'optimism')
       & (df['time'] == '2025-02-19 22:45:11')]
)
op['amount'] / 1_000_000
```

```python
(
    df[df['to_artifact_name'] == 'aave-v3']
    .groupby(['to_artifact_namespace', 'time'], as_index=False)
    ['amount']
    .sum()
    .sort_values(by='time')
).tail(29).sort_values(by='to_artifact_namespace')
```

```python
df[df['to_artifact_name'] == 'aave-v3'].groupby('time')['amount'].sum()
```

```python
(
    df[(df['to_artifact_name'] == 'aave-v3') & (df['to_artifact_namespace'] == 'optimism')].groupby('time')['amount'].sum()
).iloc[-2]
```
