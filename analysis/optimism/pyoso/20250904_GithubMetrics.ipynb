```python
from dotenv import load_dotenv
import os
from pyoso import Client

load_dotenv()
OSO_API_KEY = os.environ['OSO_API_KEY']
client = Client(api_key=OSO_API_KEY)
```

# Step 1: Choose a sample repo

```python
client.to_pandas("""
SELECT DISTINCT
  artifact_id,
  project_source,
  project_namespace,
  project_name,
  project_id
FROM artifacts_by_project_v1
WHERE
  artifact_source = 'GITHUB'
  AND artifact_namespace = 'ethereum-optimism'
  AND artifact_name = 'op-geth'
""")
```

# Step 2: See what GitHub metrics are available

```python
client.to_pandas("""
SELECT DISTINCT
  metric_name,
  display_name,
  metric_id  
FROM metrics_v0
WHERE metric_name LIKE 'GITHUB%'
""")
```

```python
client.to_pandas("""
SELECT
  display_name,
  ARRAY_AGG(metric_name) AS metric_variants,
  COUNT(distinct metric_name) AS metric_count
FROM metrics_v0
WHERE metric_name LIKE 'GITHUB%'
GROUP BY 1
ORDER BY 3 DESC
""")
```

# Step 3: Visualize some of the metrics

```python
df = client.to_pandas(f"""
  SELECT
    sample_date,
    amount
  FROM timeseries_metrics_by_project_v0
  JOIN metrics_v0 USING metric_id
  WHERE
    project_id = 'ta+k40iL4Aishk458IqTS4k+jcGiSCdYzpwaDPfKqss='
    AND metric_name = 'GITHUB_full_time_developers_monthly'
  ORDER BY sample_date
""")
df
```

```python
df.set_index('sample_date')['amount'].plot(figsize=(10,5))
```

# Step 4: See how many "downtream" teams are using a package

```python
df = client.to_pandas("""
  WITH deps AS (
    SELECT DISTINCT dependent_artifact_id
    FROM sboms_v0 -- this model has the full dependency tree :)
    WHERE package_artifact_name = 'viem'
  ),
  projects AS (
    SELECT DISTINCT project_id
    FROM artifacts_by_project_v1 AS abp
    JOIN deps ON abp.artifact_id = deps.dependent_artifact_id
    WHERE project_source = 'OSS_DIRECTORY' -- keep it simple for now!
  )
  SELECT
    sample_date,
    SUM(amount) as downstream_devs
  FROM timeseries_metrics_by_project_v0
  JOIN metrics_v0 USING metric_id
  JOIN projects USING project_id
  WHERE metric_name = 'GITHUB_full_time_developers_monthly'
  GROUP BY sample_date
  ORDER BY sample_date
""")
df.tail()
```

```python
df.set_index('sample_date')['downstream_devs'].plot(figsize=(10,5))
```
